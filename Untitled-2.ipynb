```python
import pandas as pd

df = pd.DataFrame({
    "Revenue Product 1": [500],
    "Revenue Product 2": [100],
    "Revenue Product 3": [600],
    "Revenue Product 4": [250],
    "Net revenue": [1450], 
    "Fixed Cost": [-200],
    "Variable Cost": [-400],
    "Trips Expenses": [-300],
    "Other expenses": [-10],
    "Operating profit": [540],
    "Income tax": [-200],
    "Net Profit": [340]
}, index=['Value'])

df = df.T   # transpose
print(df)
```

```python
df.plot(kind="bar", legend=False, figsize=(10,6), title="Business Financial Breakdown")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x=df.index, y="Value", data=df.reset_index())
plt.xticks(rotation=75)
plt.title("Business Financial Breakdown")
plt.grid(True)
plt.show()
```
